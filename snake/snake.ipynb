```python
import random
import msvcrt
import os


class snake(object):
    
    def __init__(self,board_length,board_height):
        self.speed = 1
        self.coord = [[int(board_length/2),int(board_height/2)]]
        self.direction = b'l'
        self.alive = True
        self.length = board_length
        self.height = board_height
        self.snake_len = 1

    def peek(self):
        x = self.coord[-1][0]
        y = self.coord[-1][1]

        if self.direction == b'j':
            return [x-1,y]
        elif self.direction == b'i':
            return [x,y-1]
        elif self.direction == b'k':
            return [x,y+1]
        elif self.direction == b'l':
            return [x+1,y]

    def move(self):
        next_coord = self.peek()
        self.coord.pop(0)

        for vector in self.coord:
            if vector == next_coord:
                self.alive = False

        self.coord.append(next_coord)
        
        if next_coord[0]*next_coord[1] < 0 or next_coord[0] >= self.length or next_coord[1] >= self.height:
            self.alive = False

    def grow(self):
        self.coord.append(self.peek())
        self.snake_len += 1

        if self.snake_len > self.length * self.height:
            self.alive = False
    
    def on_snake(self, x,y):
        for pos in self.coord:
            if [x, y] == pos:
                return True
        return False


class food(object):

    def __init__(self):
        random.seed()
        self.x = 0
        self.y = 0
    
    def move_food(self, snake):
        self.x += 1

        if self.x == snake.length:
            self.x = 0
            self.y += 1
            if self.y == snake.height:
                self.y = 0
    
    def make_food(self, snake):
        self.x = random.randint(0, snake.length-1)
        self.y = random.randint(0, snake.height-1)
        

        while snake.on_snake(self.x, self.y):
            if snake.snake_len < snake.length * snake.height:
                move_food()
            else:
                break

class screen(object):
    
    def __init__(self):
        self.head = "The Greedy Snake Game\n"
        self.body = []
        self.best_length = 1
    
    def print_header(self, snake):
        hspace = int((snake.length - 10)/2)

        if hspace > 0:
            for i in range(hspace):
                print(' ', end ='')
            print(self.head)
        else:
            print(self.head)
        


    def print_body(self, snake, food):
        print('+',end = '')
        for x in range(snake.length):
            print('-', end = '')
        print('+')

        for y in range(snake.height):
            print('|', end = '')
            for x in range(snake.length):
                if snake.on_snake(x,y):
                    if [x,y] == snake.coord[-1]:
                        print('S', end = '')
                    else:
                        print('@', end ='')
                elif [x,y] == [food.x, food.y]:
                    print('*', end = '')
                else:
                    print(' ', end = '')
            print('|')
        
        print('+',end = '')
        for x in range(snake.length):
            print('-', end = '')
        print('+')

    def print_foot(self, snake):
        print(f"Your Score: {snake.snake_len}")

    def print_scr(self, snake, food):
        self.print_header(snake)
        self.print_body(snake, food)
        self.print_foot(snake)


def snake_game():
    print("Select size of the board.")
    L = int(input("length: "))*2
    H = int(input("height: "))
    delay = int(input("set delay: ")) * 100

    Snake = snake(L,H)
    Food = food()
    Screen = screen()
    
    counter = 0

    Food.make_food(Snake)

    while Snake.alive:
        if msvcrt.kbhit():
            key = msvcrt.getch()
            if key == b'i' or key == b'j' or key == b'k' or key == b'l':
                Snake.direction = key
                Snake.move()
                counter = 0
                os.system('cls')
                Screen.print_scr(Snake, Food)
        
        if counter >= delay:
            Snake.move()
            os.system('cls')
            Screen.print_scr(Snake, Food)
            counter = 0

        if Snake.coord[-1] == [Food.x, Food.y]:
            Snake.grow()
            os.system('cls')
            Screen.print_scr(Snake, Food)
            Food.make_food(Snake)

        counter += 1

snake_game()
```
