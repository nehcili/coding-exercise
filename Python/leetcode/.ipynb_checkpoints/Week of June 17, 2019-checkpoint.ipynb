```python
class CSII1:
    def combinationSum2(self, candidates, target: int):
        self.ans = []
        candidates.sort()
        n = len(candidates)
        
        def helper(curList, candInd, target):
            if target == 0:
                self.ans.append(curList[:])
                return
            if candInd == n or target < 0:
                return
            
            curList.append(candidates[candInd])
            helper(curList, candInd + 1, target - candidates[candInd])
            curList.pop()
            
            i = candInd+1
            while i < n and candidates[candInd] == candidates[i]:
                i += 1
            
            helper(curList, i, target)
            
        helper([], 0, target)
        return self.ans
```

```python
class CSII2:
    def combinationSum2(self, candidates, target: int):
        def dfs(candidates, target, s, curr, ans):
            if target == 0: 
                ans.append(curr[:])
                return
            
            for i in range(s, len(candidates)):
                if candidates[i] > target:
                    return
                else:
                    if i > s and candidates[i] == candidates[i-1]:
                        continue
                    curr.append(candidates[i])
                    dfs(candidates, target - candidates[i], i+1, curr, ans)
                    curr.pop()
                
        curr = []
        ans = []
        candidates.sort()
        dfs(candidates, target, 0, curr, ans)
        return ans

x = CSII()
```

```python
from collections import Counter
class WS2:
    def exist(self, board, word: str) -> bool:
        count=Counter(x for rows in board for x in rows)
        for c, cnt in Counter(word).items():
            if cnt>count[c]:return False
        r= len(board)
        c=len(board[0])
        length=len(word)
        def go(x,y,index):
            if index==length:
                return True
            w=word[index]
            prev=board[x][y]
            board[x][y]=''
            if x>0 and board[x-1][y]==w and go(x-1,y,index+1):
                return True
            if x+1<r and board[x+1][y]==w and go(x+1,y,index+1):
                return True
            if y>0 and board[x][y-1]==w and go(x,y-1,index+1):
                return True
            if y+1<c and board[x][y+1]==w and go(x,y+1,index+1):
                return True
            board[x][y]=prev
            return False
        candidates=((x,y) for x in range(r) for y in range(c) if board[x][y]==word[0])
        return any(go(x,y,1) for x,y in candidates)
```

    - DW1 and DW2 use the same idea. Though DW2 is much smarter in implementation. Note that the dict in DW1 is really not needed. All that is needed at each stage n is number of ways to decode for n-1 and n-2. This is implemented in DW2 as dp2 and dp1, respectively.

```python
class BCII1:
    def calculate(self, s: str) -> int:
        slist = []
        i = 0
        n = len(s)
        
        while i < n:
            if s[i] == ' ':
                i += 1
            elif s[i] in '+-*/':
                slist.append(s[i])
                i += 1
            else:
                j = i + 1
                while j < n and s[j] in '0123456789':
                    j += 1
                slist.append(int(s[i:j]))
                i = j
        
        #print(slist)
        ans = []
        i = 0
        while i < len(slist):
            if slist[i] == '*':
                ans[-1] *= slist[i+1]
                i += 2
            elif slist[i] == '/':
                ans[-1] = int(ans[-1]/slist[i+1])
                i += 2
            elif slist[i] == '-':
                slist[i+1] *= -1
                i += 1
            elif slist[i] == '+':
                i += 1
            else:
                ans.append(slist[i])
                i += 1
        
        
        
        return sum(ans)
```

```python
class BCII2:
    def calculate(self, s: str) -> int:
        s += '+'
        stack = []
        num = 0
        operator = '+'
        
        for l in s:
            if l in '0123456789':
                num = num*10 + int(l)
            elif l != ' ':
                if operator == '+':
                    stack.append(num)
                elif operator == '-':
                    stack.append(-num)
                elif operator == '*':
                    stack.append(stack.pop() * num)
                else:
                    stack.append(int(stack.pop()/num))
                operator = l
                num = 0
            else:
                continue
        
        return sum(stack)
            
            
            
                
x = BCII2()                
```

```python
# BC
class BC:
    def calculate(self, s: str) -> int:
        s += '+'
        stack = []
        signs = []
        num = 0
        operator = '+'
        
        for l in s:
            if l in '0123456789':
                num = num*10 + int(l)
            elif l == '(':
                temp = 1 if operator == '+' else -1
                if signs:
                    signs.append(temp * signs[-1])
                else:
                    signs.append(temp)
                operator = '+'
            elif l != ' ':
                if signs:
                    sign = signs[-1]
                else:
                    sign = 1
                #print(signs, num, operator)
                    
                if operator == '+':
                    stack.append(sign*num)
                elif operator == '-':
                    #print('sdfsdf',-num*sign)
                    stack.append(-num*sign)
                
                
                if l == ')':
                    signs.pop()
                else:
                    operator = l
                num = 0
            else:
                continue
        
        return sum(stack)
        
x = BC()        
```

```python
class DW1:
    def numDecodings(self, s: str) -> int:
        if s == '0':
            return 0
        dict = {}
        n = len(s)
        
        def helper(ind):
            if ind == n-1:
                dict[ind] = int(int(s[ind]) > 0)
            elif ind == n-2:
                if int(s[ind]) > 0 and int(s[ind+1]) > 0 and int(s[ind:ind+2]) < 27:
                    dict[ind] = 2
                elif int(s[ind]) > 0 and int(s[ind:ind+2]) < 27:
                    dict[ind] = 1
                elif int(s[ind]) > 0 and int(s[ind+1]) > 0:
                    dict[ind] = 1
                else:
                    dict[ind] = 0
            elif ind in dict:
                return dict[ind]
            elif int(s[ind]) == 0:
                dict[ind] = 0
            elif int(s[ind+1]) > 0 and int(s[ind:ind+2]) < 27:
                dict[ind] = helper(ind+1) + helper(ind+2)
            elif int(s[ind:ind+2]) < 27:
                dict[ind] = helper(ind+2)
            else:
                dict[ind] = helper(ind+1)
            
            return dict[ind]
            
        
        return helper(0)
            
            
        
```

```python
class DW2:
    def numDecodings(self, s: str) -> int:
        
        dic = { str(i-64):chr(i) for i in range( ord('A'), ord('Z') + 1) }
        # print(dic)
        # dp
        dp1 = 1
        dp2 = 1
        
        
        for i in range(len(s)):
            dp_new = 0
            # print(s[i]) 
            if s[i] in dic:
                dp_new += dp2
                
            if i>0 and s[(i-1):(i+1)] in dic:
                # print(s[(i-1):(i+1)])
                dp_new += dp1
            
            dp1 = dp2
            dp2 = dp_new
        
        return dp2
```

- kth smallest element in a BST
    - kSEBST1: my recurssion
    - kSEBST2: pro's non recursive solution
- Binary Tree Inorder Traversal
    - recursive solution is trivial. Iterative solution is the same as kSEBST2. Both solution uses O(n) space
    - Morris Traversal - if a root has a left child, put root as the right child of the right most child of root.left. time O(n) and space O(1)
- Valid Binary Search Tree
- Climbing Stairs
- Populating Next Right Pointers in Each Node
- Word Ladder
    - WL1: me
    - WL2: a bit better. it takes O(n^2) to make the graph, but can do much faster if we just add all possible neighbors. word[:i] + '*' + word[i+1:]

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class kSEBST1:
    def kthSmallest(self, root: TreeNode, k: int) -> int:
        def helper(subroot, size):
            if subroot == None:
                return (False, size)
            elif subroot.left == None and k == size+1:
                return (True, subroot.val)
            elif subroot.left == None and k > size + 1:
                return helper(subroot.right, size+1)
            
            a,b = helper(subroot.left, size)
            if a:
                return (True, b)
            elif b+1 == k:
                return (True, subroot.val)
            else:
                return helper(subroot.right,  b+1)
        
        return helper(root, 0)[1]
```

```python
from collections import deque
class Solution:
    def kthSmallest(self, root: TreeNode, k: int) -> int:
        q = [(root, True)]
        while q:
            node, explore = q.pop()
            if node:
                if explore:
                    q.append((node.right, True))
                    q.append((node, False))
                    curr = node.left
                    while curr:
                        q.append((curr.right, True))
                        q.append((curr, False))
                        curr = curr.left
                else:
                    k -= 1
                    if k == 0:
                        return node.val
```

```python
class BTIT:
    def inorderTraversal(self, root: TreeNode):
        res = []
        
        q = [(root, True)]
        while q:
            node, explore = q.pop()
            if node:
                if explore:
                    q.append((node.right, True))
                    q.append((node, False))
                    cur = node.left
                    while cur:
                        q.append((cur.right, True))
                        q.append((cur, False))
                        cur = cur.left
                else:
                    res.append(node.val)
        
        return res
```

```python
class WL1:
    def ladderLength(self, beginWord: str, endWord: str, wordList) -> int:
        def isNhbr(a,b):
            diff = 0
            for i in range(len(a)):
                if a[i] != b[i]:
                    diff += 1
                    if diff > 1:
                        return False
            return True
        
        n = len(wordList)
        res = 0
        q = [beginWord]
        while q:
            res += 1
            temp = []
            #print(res, q)
            while q:
                word = q.pop()
                if word == endWord:
                    return res
                for i in range(n):
                    if isNhbr(word, wordList[i]):
                        temp.append(wordList[i])
                        wordList[i] = '!!'
            if temp == []:
                return 0
            q = temp
            
        return res
    
x = WL1()
```

```python
from collections import defaultdict
from collections import deque
class WL2:
    def ladderLength(self, beginWord: str, endWord: str, wordList) -> int:
        L = len(beginWord)
        G = defaultdict(list)
        for word in wordList:
            for i in range(L):
                G[word[:i]+'*'+word[i+1:]].append(word)
        
        q = deque([(beginWord,1)])
        visited = set()
        while q:
            word, level = q.popleft()
            if word == endWord:
                return level
            level +=1 
            
            for i in range(L):
                prefix, suffix = word[:i], word[i+1:]
                if prefix+'*'+suffix in G:
                    for nhbr in G[prefix+'*'+suffix]:
                        if nhbr not in visited:
                            q.append((nhbr, level))
                            visited.add(nhbr)
         
        return 0
```
