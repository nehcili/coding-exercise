```python
# brute foce version

def twoSum_1(nums, target):
    length = len(nums)
    
    for i in range(0, length):
        if i == length - 2:
            return [i, length-1]
        else:
            for j in range(i+1, length):
                if target == nums[i] + nums[j]:
                    return [i, j]
```

```python
# slightly better brute force version
def twoSum_2(nums, target):
        i = 0
        j = 1
        
        while True:
            while True:
                try:
                    a = nums[j]
                except IndexError:
                    j = i + 2
                    break
                
                if target == nums[i] + a:
                        return [i, j]
                j += 1
            
            i+= 1
```

```python
# new attempt Sept 10, 2018: This is accepted by Leetcode run time is O(n). OrderedDict is a hashtable, so O(1) look up time.

from collections import OrderedDict

#class Solution:
def twoSum(nums, target):
    d = OrderedDict()
    i = 0

    while True:
        # print(d)
        try:
            a = d[nums[i]]
            a.append(i)
            i += 1
        except IndexError:
            break
        except KeyError:
            d[nums[i]] = [i]
            i += 1

    for key in d:
        try:
            j = d[target - key]
            i = d[key]
            if key != target - key:
                return [i[0],j[0]]
            elif len(i) >= 2:
                return [i[0], i[1]]
            else:
                continue
        except KeyError:
            continue

```
