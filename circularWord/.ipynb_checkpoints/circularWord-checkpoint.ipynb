```python
# The excercise:
# A list of words is called circular if they can be arranged in an order such that,
# 1) (except for the first word) for each word, its first letter is equal to the last
#    letter of the word in front of it
# 2) the last letter of the last word is equal to the first letter of the first word
#
# Given a list of words, determine if it is circular.
#
# Idea of solution: 
# We note that if a list is circular, it doesn't matter which word is the first word

class state(object):
    def __init__(self, node, nbhd, rest):
        self.node = node
        self.nbhd = nbhd
        self.rest = rest

    def hasNb(self):
        return self.nbhd != []

    def nextNbhd(self):
        while self.hasNb():
            if self.node[-1] == self.nbhd[-1][0]:
                return self.nbhd[:]
            self.nbhd.pop()
        return False

    def nextNb(self):
        self.nextNbhd()
        if self.hasNb():
            return self.nbhd[-1]
        else:
            return False

    def popNextNb(self):
        self.nextNbhd()
        if self.hasNb():
            self.nbhd.pop()

    def restRemove(self, ANode):
        n = self.rest[:]
        n.remove(ANode)
        return n

    def nextState(self):
        newNbhd = self.nextNbhd()

        if newNbhd != False:
            newNode = newNbhd[-1]
            newNbhd.pop()
            return state(newNode, self.restRemove(newNode), self.restRemove(newNode))
        else:
            return False

    def print_state(self):
        print(self.node)
        print(self.nbhd)
        print(self.rest)

def hasCircularWord(lstOfStr):
    l = len(lstOfStr)
    
    # base case when the length of lstOfStr is 1
    if l == 1:
        if lstOfStr[0] == '':
            return False
        return lstOfStr[0][0] == lstOfStr[0][-1]

    for word in lstOfStr:
        lst = lstOfStr[:]
        lst.remove(word)
        lst2 = lst[:]
        circle = [state(word, lst, lst2)]
        i = 0

        while circle != []:

            newNb = circle[i].nextNb()
            if newNb != False:
                circle.append(circle[i].nextState())
                i += 1
                #for s in circle:
                #    s.print_state()
            elif (i == l-1) & (circle[-1].node[-1] == circle[0].node[0]):
                return True
            elif i == 0:
                break
            else:
                i -= 1
                if circle[i].nextNb() != False:
                    circle[i].popNextNb()
                else:
                    circle.pop()
    
    return False
```
